```python
# Load models at the start
model_dict = {}
ions = ['Cl', 'Br', 'Mg', 'Na', 'TDS', 'Ca', 'SO4', 'K']
for ion in ions:
    model_dict[ion] = keras.models.load_model(f'Models/ANN_{ion}.h5')
```

```python


def Ion_Simulator(EC, Sacramento_X2, Ion, WYT, Region, month, model):
    X = pd.DataFrame(columns=['EC', 'Sacramento X2', 'AN', 'BN', 'C', 'D', 'W', 'OMR', 'SJRcorridor', 'SouthDelta', 'April', 'August', 'December', 'February', 'January', 'July', 'June', 'March', 'May', 'November', 'October', 'September'])
    X.loc[0,"EC"]=(EC -50)/(3500-50)
    X.loc[0,'Sacramento X2']=Sacramento_X2/100
    X.loc[0,'AN']=1 if WYT=="AN" else 0
    X.loc[0,'BN']=1 if WYT=="BN" else 0
    X.loc[0,'C']=1 if WYT=="C" else 0
    X.loc[0,'D']=1 if WYT=="D" else 0
    X.loc[0,'W']=1 if WYT=="W" else 0

    X.loc[0,'OMR']=1 if Region=="OMR" else 0
    X.loc[0,'SJRcorridor']=1 if Region=="SJRcorridor" else 0
    X.loc[0,'SouthDelta']=1 if Region=="SouthDelta" else 0

    X.loc[0,'April']=1 if month=="April" else 0
    X.loc[0,'August']=1 if month=="August" else 0
    X.loc[0,'December']=1 if month=="December" else 0
    X.loc[0,'February']=1 if month=="February" else 0
    X.loc[0,'January']=1 if month=="January" else 0
    X.loc[0,'July']=1 if month=="July" else 0
    X.loc[0,'June']=1 if month=="June" else 0
    X.loc[0,'March']=1 if month=="March" else 0
    X.loc[0,'May']=1 if month=="May" else 0
    X.loc[0,'November']=1 if month=="November" else 0
    X.loc[0,'October']=1 if month=="October" else 0
    X.loc[0,'September']=1 if month=="September" else 0

    ANN_final= keras.models.load_model(f'Models/ANN_{Ion}.h5')

    X1 = X.astype(float)
    df = pd.DataFrame(columns=['ANN'])
    df.loc[0, 'ANN'] = round(model.predict(X1, verbose=0)[0][0].astype(float), 2)
    result = df

    return result


    
```

```python
def process_data(specific_date, selected_ion, model):
    # Filter 'df_merged' to include only the rows for the specific date
    df_filtered_OM = df_merged_OM[df_merged_OM['Date'] == pd.to_datetime(specific_date)]
    df_filtered_SJ = df_merged_SJ[df_merged_SJ['Date'] == pd.to_datetime(specific_date)]
    df_filtered_SD = df_merged_SD[df_merged_SD['Date'] == pd.to_datetime(specific_date)]

    # Prepare lists to collect the simulation DataFrames for each region
    simulation_results_OM = []
    simulation_results_SJ = []
    simulation_results_SD = []
     # Function to run the simulation for a given DataFrame, location range, and region
    def run_simulation(df_filtered, location_prefix, location_range, region, model):
        simulation_results = []
        for location in [location_prefix + str(i).zfill(2) for i in range(1, location_range + 1)]:
            location_results = []
            for index, row in df_filtered.iterrows():
                month = row['Date'].strftime('%B')
                result = Ion_Simulator(
                    EC=row[location],          
                    Sacramento_X2=row['X2'],   
                    Ion=selected_ion,                  
                    WYT=row['WYT'],            
                    Region=region,             
                    month=month,
                    model=model  # Pass the pre-loaded model here
                )
                result['Date'] = row['Date']
                result['Location'] = location
                location_results.append(result)
            simulation_results.append(pd.concat(location_results, ignore_index=True))
        return pd.concat(simulation_results, ignore_index=True)
    
    # Run simulation for each DataFrame
    df_simulations_OM = run_simulation(df_filtered_OM, 'OM', 72, 'OMR', model)
    df_simulations_SJ = run_simulation(df_filtered_SJ, 'SJ', 30, 'SJRcorridor', model)
    df_simulations_SD = run_simulation(df_filtered_SD, 'SD', 60, 'SouthDelta', model)


    # Pivot the DataFrame
    df_pivoted_OM = df_simulations_OM.pivot(index='Date', columns='Location', values='ANN')
    df_pivoted_SJ = df_simulations_SJ.pivot(index='Date', columns='Location', values='ANN')
    df_pivoted_SD = df_simulations_SD.pivot(index='Date', columns='Location', values='ANN')
    # Reset the index to make 'Date' a column again
    df_pivoted_OM.reset_index(inplace=True)
    df_pivoted_SJ.reset_index(inplace=True)
    df_pivoted_SD.reset_index(inplace=True)

    # Rename the column level
    df_pivoted_OM.columns.name = None
    df_pivoted_SJ.columns.name = None
    df_pivoted_SD.columns.name = None

    df_OM =df_pivoted_OM
    df_SJ =df_pivoted_SJ
    df_SD =df_pivoted_SD

    cols_to_convert = df_OM.columns.drop('Date')  # Exclude 'Date' from conversion
    df_OM[cols_to_convert] = df_OM[cols_to_convert].apply(pd.to_numeric, errors='coerce')

    cols_to_convert = df_SJ.columns.drop('Date')  # Exclude 'Date' from conversion
    df_SJ[cols_to_convert] = df_SJ[cols_to_convert].apply(pd.to_numeric, errors='coerce')

    cols_to_convert = df_SD.columns.drop('Date')  # Exclude 'Date' from conversion
    df_SD[cols_to_convert] = df_SD[cols_to_convert].apply(pd.to_numeric, errors='coerce')


    # Convert 'Date' columns to string type
    df_SJ['Date'] = df_SJ['Date'].astype(str)
    df_SD['Date'] = df_SD['Date'].astype(str)
    df_OM['Date'] = df_OM['Date'].astype(str)

    # Assuming all columns except 'Date' contain relevant values
    values_OM = df_OM.drop('Date', axis=1).values.flatten()
    values_SJ = df_SJ.drop('Date', axis=1).values.flatten()
    values_SD = df_SD.drop('Date', axis=1).values.flatten()

    # Combine all values and calculate global min and max
    all_values = np.concatenate([values_OM, values_SJ, values_SD])
    GLOBAL_MIN_VAL = np.nanmin(all_values)
    GLOBAL_MAX_VAL = np.nanmax(all_values)


    # Function for linear interpolation
    def linear_interpolate(start, end, num_points=5):
        return np.linspace(start, end, num_points)

    # Function to map a value to a color
    def value_to_color(value, min_val, max_val):
        norm = plt.Normalize(vmin=min_val, vmax=max_val)
        cmap = plt.cm.viridis
        rgba = cmap(norm(value))
        return mcolors.rgb2hex(rgba)

    # Modified function to add points and interpolated lines to the map
    def add_interpolated_lines_to_map(df, lat_long, prefix, map_obj, line_weight=15):
        for i in range(len(lat_long['lat']) - 1):
            start_lat, start_lon = lat_long['lat'][i], lat_long['lon'][i]
            end_lat, end_lon = lat_long['lat'][i+1], lat_long['lon'][i+1]
            start_value = df[prefix + str(i+1).zfill(2)].iloc[0]
            end_value = df[prefix + str(i+2).zfill(2)].iloc[0]

            # Interpolate points and values
            interpolated_lats = linear_interpolate(start_lat, end_lat)
            interpolated_lons = linear_interpolate(start_lon, end_lon)
            interpolated_values = linear_interpolate(start_value, end_value)

            min_val, max_val = min(interpolated_values), max(interpolated_values)

            # Create line segments
            for j in range(len(interpolated_lats) - 1):
                segment = [(interpolated_lats[j], interpolated_lons[j]), 
                        (interpolated_lats[j+1], interpolated_lons[j+1])]
                color = value_to_color(interpolated_values[j], GLOBAL_MIN_VAL, GLOBAL_MAX_VAL)
                folium.PolyLine(segment, color=color, weight=line_weight).add_to(map_obj)

    # Create a base map
    m = folium.Map(location=[37.9, -121.5], zoom_start=11, width='100%', height='100%')

    # Apply the function to your sets
    for set_index in range(0, 9):
        if set_index == 0:
            add_interpolated_lines_to_map(df_SJ, lat_long_sets[set_index], 'SJ', m)
        elif 1 <= set_index <= 5:
            add_interpolated_lines_to_map(df_SD, lat_long_sets[set_index], 'SD', m)
        elif 6 <= set_index <= 8:
            add_interpolated_lines_to_map(df_OM, lat_long_sets[set_index], 'OM', m)

            
    colormap = branca.colormap.LinearColormap(
        colors=[plt.cm.viridis(i) for i in range(256)],
        index=np.linspace(GLOBAL_MIN_VAL, GLOBAL_MAX_VAL, 256),
        vmin=GLOBAL_MIN_VAL,
        vmax=GLOBAL_MAX_VAL
    )

    colormap.caption = "Ion Level"

    # Add the color scale bar to your map
    colormap.add_to(m)

    # Custom HTML for styling the color scale bar
    custom_html = """
    <style>
        .folium-map .leaflet-control .leaflet-bar a {
            font-size: 72px; /* Increase the font size of the color scale numbers */
        }
        .folium-map .leaflet-control .leaflet-bar {
            width: 200px; /* Increase the width of the color bar */
        }
    </style>
    """

    # Adding custom HTML to the map
    m.get_root().html.add_child(folium.Element(custom_html))
    

    # HTML to be displayed in the iFrame
    html = f"""
    <!DOCTYPE html>
    <html>
    <head>
    <title>Map Title</title>
    <style>
        .title {{
            position: fixed;
            top: 10px;
            left: 50%;
            transform: translateX(-50%);
            background-color: rgba(255, 255, 255, 0.7);
            padding: 5px 10px;
            border-radius: 5px;
            font-size: 20px;
            font-weight: bold;
        }}
    </style>
    </head>
    <body>
    <div class="title">{selected_ion} Level on {specific_date}</div>
    </body>
    </html>
    """

    # Create an iFrame using the HTML
    iframe = IFrame(html=html, width=300, height=80)

    # Add the iFrame to the map at a specific location
    folium.Marker(
        location=[38.1, -121.5],  # You can choose a suitable location
        icon=folium.DivIcon(html=iframe.render())
    ).add_to(m)




    m.save('map.html')

    # Display the map
    return m
    
    # Function to handle date change
def on_date_change(change):
    clear_output(wait=False)
    display(date_picker)
    display(ion_picker)
    selected_ion = ion_picker.value
    ion_model = model_dict[selected_ion]
    m = process_data(change['new'], selected_ion, ion_model)  # Pass the model here
    display(m)

# Function to handle ion change
def on_ion_change(change):
    clear_output(wait=False)
    display(date_picker)
    display(ion_picker)
    selected_ion = change['new']
    ion_model = model_dict[selected_ion]
    m = process_data(date_picker.value, selected_ion, ion_model)  # Pass the model here
    display(m)

# Set up the date picker widget
date_picker = widgets.DatePicker(
    description='Select Date',
    value=pd.to_datetime('2015-07-01')  # Default date
)

# Create the ion selection dropdown
ion_picker = widgets.Dropdown(
    options=ions,
    value='Cl',  # default value
    description='Select Ion:',
)

# Link the widgets to their respective handlers
date_picker.observe(on_date_change, names='value')
ion_picker.observe(on_ion_change, names='value')

# Display the widgets
display(date_picker)
display(ion_picker)

# Generate and display the initial map
selected_ion = ion_picker.value
ion_model = model_dict[selected_ion]
initial_map = process_data(date_picker.value, selected_ion, ion_model)  
display(initial_map)
```
